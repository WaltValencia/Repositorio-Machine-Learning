Se crea la variable _numeros_ con lo encontrado anteriormente
 

```python
numeros = pag_revista_sopa.find_all('ul', attrs={'class':'issues_archive'})
```

Para el siguiente comando, se utiliza un for para decirle al colab cuáles son los datos específicos de cada artículo que se necesitan, en este caso, utilizamos el título del artículo, los autores, el resumen (abstract) y la edición (volumen de la revista).

```python
datos = {'titulo':[],
         'autores':[],
         'abstract':[],
         'edicion':[]}
contador = 0
for numero in numeros:
  pag_numero = rq.get(numero.a.get('href'))
  pag_numero_sopa = BeautifulSoup(pag_numero.text, 'lxml')
  articulos = pag_numero_sopa.find_all('h3', attrs={'class':'title'})
  for articulo in articulos:
    pag_articulo = rq.get(articulo.a.get('href'))
    pag_articulo_sopa = BeautifulSoup(pag_articulo.text, 'lxml')
    datos ['titulo'].append(pag_articulo_sopa.find('h2', attrs={'class':'page_title'}).text)
    datos ['autores'].append(pag_articulo_sopa.find('span', attrs={'class':'name'}).text)
    datos ['abstract'].append(pag_articulo_sopa.find('div', attrs={'class':'tab-pane active'}).text)
    datos ['edicion'].append(pag_articulo_sopa.find('div', attrs={'class':'ruta'}).find_all('a')[3].text)
    contador+=1
    print(contador)
```

Como resultado, se encuentran 13 artículos que están contenidos en pag_articulo_sopa

El comando contador permite contar el numero de operaciones que se estan realizando

Se importan las librerías pandas para crear un dataframe y convertir el archivo en un csv

```python
import pandas as pd

data_texto = pd.DataFrame(datos)
data_texto.to_csv('data_texto.csv')
```

# Lowering

Para no tener problemas de limpieza de datos más adelante, se hace uso de la opción .lower, que permite poner todo el texto en minúscula

```python
texto_lower = texto_inicial.lower()
```

```python
texto_lower
```

Expresiones regulares

Se quita el \na del inicio del texto y el \t del final, con el fin de hacer una limpieza mejor:

```python
texto_corr = texto_lower.replace('\n',' ').replace('\t',' ').strip()
```

```python
texto_corr
```

Tokenización

Con la tokenización, se divide el artículo en palabras y se contienen en una lista, llamada texto_toke

```python
texto_toke = texto_corr.split()
texto_toke
```

```python
import nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt')

texto_token = word_tokenize(texto_corr)
```
