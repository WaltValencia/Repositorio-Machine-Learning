##**INCIDENCIA EN EL NIVEL DE ESCOLARIDAD EN COLOMBIA**##

De acuerdo con diversas investigaciones por universidades de prestigio y entidades gubernamentales tanto de Colombia como otros paises, se puede definir el nivel de escolaridad primaria depende directamente del gasto pùblico en educación de los paises

## Primer paso ##

Se extrae la base de datos de la pagina oficial del Banco Mundial, sin tener que descargarla, directamente desde el api y se abre para observar que todos los datos esten cargados correctamente, como no se necesitan todos los datos que suministra la pagina, se procede a filtrar desde el API solo para Colombia.

```python
#Importar librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import requests
from sklearn.linear_model import LinearRegression

#Traer api Banco Mundial - Variable: Gasto Público en Educación

dataset = ('https://api.worldbank.org/v2/country/COL/indicator/SE.XPD.TOTL.GB.ZS?'
          'per_page=61'
          '&format=json'
          '&date=1998:{}'. format(2021)  
           )


response = requests.get(dataset)


payload = response.json()

df_Educ_Expen = pd.DataFrame.from_dict(payload[1])

df_Educ_Expen
```

## Segundo Paso ## 

Luego de eso se observa que la primera base de datos correspondiente al gasto en educaciòn de Colombia esta cargada correctamente y se procede a subir las otras tres bases de datos.

```python
#Traer api Banco Mundial - Variable: Inscripción escolar: nivel primario (% bruto)

dataset1 = ('https://api.worldbank.org/v2/country/COL/indicator/SE.PRM.ENRR?'
          'per_page=61'
          '&format=json'
          '&date=1998:{}'. format(2021)  
           )

response1 = requests.get(dataset1)


payload1 = response1.json()

df_Edu_prima = pd.DataFrame.from_dict(payload1[1])

df_Edu_prima
```

```python
#Traer api Banco Mundial - Variable: Población Total

dataset2 = ('https://api.worldbank.org/v2/country/COL/indicator/SP.POP.TOTL?'
          'per_page=61'
          '&format=json'
          '&date=1998:{}'. format(2021)  
           )

response2 = requests.get(dataset2)


payload2 = response2.json()

df_poblacion = pd.DataFrame.from_dict(payload2[1])

df_poblacion
```

```python
#Traer api Banco Mundial - Variable: PIB Per Capita

dataset3 = ('http://api.worldbank.org/v2/country/COL/indicator/NY.GDP.PCAP.CD?' 
          'per_page=61'
          '&format=json'
          '&date=1998:{}'. format(2021)  
           )
response3 = requests.get(dataset3)


payload3 = response3.json()

df_GDP = pd.DataFrame.from_dict(payload3[1])

df_GDP
```

## Tercer paso ##

Luego de haber cargado correctamente las bases de datos, se procede a unir todas para obtener una sola base madre.

```python
df_merge_one = pd.merge(df_Educ_Expen, df_Edu_prima, how='left', left_on=['date', 'countryiso3code'], right_on=['date', 'countryiso3code'])
df_merge_one = df_merge_one.rename(columns={'value_x': 'Educ_Expen', 'value_y': 'Edu_prima'})

df_merge_two = pd.merge(df_poblacion, df_GDP, how='left', left_on=['date', 'countryiso3code'], right_on=['date', 'countryiso3code'])
df_merge_two = df_merge_two.rename(columns={'value_x': 'poblacion', 'value_y': 'GDP'})

df = pd.merge(df_merge_one, df_merge_two, how='left', left_on=['date', 'countryiso3code'], right_on=['date', 'countryiso3code'])

df
```

#**ESPECIFICACIÓN DEL MODELO**#
Teniendo en cuenta la información anterior, el modelo busca medir la relación existente entre los indicadores de inscripción de la eduación primaria con los indicadores de población, PIB Per Cápita (GDP) y Gasto Público en Educación. Es importante mencionar que la variable dependiente es la inscripción escolar, nivel primario (% en bruto) y por ende las demas variables seran las independietnes, obteniendo un modelo de la siguiente forma:

$EduPrima = \beta X_1 \beta_2 X^2_2 \beta_3 X^3_3$

**-PIB PER CÁPITA(GDP):** Esta variable se puede considerar como una de las principales para el analisis del modelo, dado que esto puede determinar el acceso a la escolaridad primaria en Colombia.

**-POBLACIÓN:**La población puede determinar la capacidad de acceso que cuenta Colombia para la educación primaria. Por lo tanto, se incluye dicha variable como una variable independiente en la especificación del modelo.

**-GASTO PÚBLICO EN EDUCACIÓN:**El Gasto Público en Educación determina el presupuesto que realiza el Gobierno Colombiano en este factor, determinando las oportunidades de acceso a la escolaridad primaria.

**-INSCRIPCIÓN ESCOLAR, NIVEL PRIMARIO (Edu_Prima):** De acuerdo con el Banco Mundial, esta variable esta determina en porcentaje en bruto. Es decir, los datos presentados estan en porcentaje que determina un aumento o disminución en la inserción escolar (primaria).

##Regresión Multiple por OLS##
Inicialmente se realiza la regresión multiple por OLS con el fin de establecer y determinar las variables y dar una breve introducción y explicación del modelo inicial. El R^2 especifica el porcentaje de varianza explicada del modelo, teniendo como resultado un 64%. 

```python
import statsmodels.api as sm
from scipy import stats

X = df.iloc[:,[10,16,22]]
y = df.Educ_Expen

X2 = sm.add_constant(X)
est = sm.OLS(y, X2)
est2 = est.fit()
print(est2.summary())
```

##Analisis##

De acuerdo con el metodo OLS (Minimos Cuadrados Ordinarios), se tiene como resultado un nivel de significancia menor al 5%. Con relación a las variables independientes, se puede establecer que no son signifacativas al modelo propuesto. Sin embargo, este proyecto pretende dar una introducción a la problematica que inicialmente se planteo, siguiendo los pasos de diversas investigaciones con el fin de presentar una mayor veracidad al modelo.

## Regresión Lasso ##

Tenemos la variable Gasto en educaciòn como la variable dependiente y la independiente las matricula de los niños en grados de primaria

```python
import time
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_squared_error, r2_score

y = df["Educ_Expen"]
lasso_model = Lasso(alpha = 0.5)
start_time = time.process_time()
feature_cols = ["Edu_prima"]
X_Lasso_uno = df.loc[:, feature_cols]


lasso_model.fit(X_Lasso_uno, y)
y_pred = lasso_model.predict(X_Lasso_uno)

print('Modelo lasso')
print('Los coeficientes son', lasso_model.coef_)
print('Los interceptos son', lasso_model.intercept_)
print('El tiempo de ejecución es', time.process_time() - start_time, 'segundos')
print('El error cuadratico dentro de muestra es', r2_score(y, y_pred))
print('El error cuadratico fuera de muestra es', mean_squared_error(y, y_pred))
print(y_pred)

plt.scatter(X_Lasso_uno, y, color = "green")
plt.title("Regresiòn Lasso Gasto en Educaciòn con Inscripciòn pre escolar")
plt.plot(X_Lasso_uno, y_pred, color = "magenta")
```

En este caso, el gráfico no demuestra que exista una correlación entre la variable Gasto Público en Educación e Inscripción escolar en nivel primaria, dado que los puntos están muy dispersos y no relacionan un orden específico. Además, la ralición esperada es positiva, dado que se espera que cuanto mayor Gasto en Educación exista, el nivel de inscripciones escolares sea más grande, abriendo las posibilidades a los niños de acceder a una educación.
